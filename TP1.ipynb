```python
import newspaper

languages = newspaper.languages()
print(languages)
```

```python
from newspaper import build

# Créer une source depuis l'URL en utilisant la langue 'ar' (arabe)
hespress_source = build("https://www.hespress.com", language='ar')

# Afficher une description de la source créée
hespress_source.print_summary()
```

```python
hespress_source.articles
```

```python
# Parcourir les articles de la source et afficher les URLs des articles
for article in hespress_source.articles:
    print(article.url)
```

```python
# Accéder au premier article de la liste
hes_article = hespress_source.articles[0]

# Afficher le titre de l'article
print("Titre de l'article :", hes_article.title)

# Afficher le contenu de l'article
print("Contenu de l'article :", hes_article.text)

# Afficher les auteurs de l'article
print("Auteurs de l'article :", hes_article.authors)
```

```python
import wikipedia

# Effectuer une recherche sur "informatique"
search_results = wikipedia.search("informatique")
print(search_results)
```

```python
# Changer la langue de recherche en arabe
wikipedia.set_lang("ar")

# Effectuer une recherche sur "informatique" en arabe
search_results_arabic = wikipedia.search("informatique")
print(search_results_arabic)
```

```python

# Changer la langue de recherche en anglais
wikipedia.set_lang("en")

# Obtenez la première page de la liste search_results
list_eng = wikipedia.search("informatique")
page = wikipedia.page(list_eng[0])

# Récupérer le titre de la page
title = page.title

# Récupérer le contenu de la page
content = page.content

# Récupérer l'URL de la page
url = page.url

# Afficher les résultats
print("Titre de la page :", title)
print("Contenu de la page :", content)
print("URL de la page :", url)
```
